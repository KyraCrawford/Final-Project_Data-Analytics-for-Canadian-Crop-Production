## Problem 1

#### Create tables

Establish a connection **conn** to the RSQLite database **FinalDB.sqlite**, and create the following four tables.

1.  **CROP_DATA**
2.  **FARM_PRICES**
3.  **DAILY_FX**
4.  **MONTHLY_FX**

The previous practice lab will help you accomplish this.


```r
# Write your query here
conn <- dbConnect(RSQLite::SQLite(),"FinalDB.sqlite")
# creating CROP_DATA table in database
df1 <- dbExecute(conn, "CREATE TABLE CROP_DATA (
CD_ID INTEGER NOT NULL, YEAR DATE NOT NULL, CROP_TYPE VARCHAR(50) NOT NULL, GEO VARCHAR(50) NOT NULL, SEEDED_AREA INTEGER NOT NULL, HARVESTED_AREA INTEGER NOT NULL, PRODUCTION INTEGER NOT NULL, AVG_YIELD INTEGER NOT NULL, PRIMARY KEY (CD_ID))")

# creating FARM_PRICES table in database
df2 <- dbExecute(conn, "CREATE TABLE FARM_PRICES (
CD_ID INTEGER NOT NULL, DATE DATE NOT NULL, CROP_TYPE VARCHAR(50) NOT NULL, GEO VARCHAR(50) NOT NULL, PRICE_PRERMT DECIMAL(6,2) NOT NULL, PRIMARY KEY (CD_ID))")

# creating DAILY_FX table in database
df3 <- dbExecute(conn, "CREATE TABLE DAILY_FX (
DFX_ID INTEGER NOT NULL, DATE DATE NOT NULL, FXUSDCAD DECIMAL(6,4) NOT NULL, PRIMARY KEY (DFX_ID))")

# creating MONTHLY_FX table in database
df4 <- dbExecute(conn, "CREATE TABLE MONTHLY_FX (
DFX_ID INTEGER NOT NULL, DATE DATE NOT NULL, FXUSDCAD DECIMAL(10,9) NOT NULL, PRIMARY KEY (DFX_ID))")
```

```r
#check list of tables in the present db.
dbListTables(conn)
```

## Problem 2

#### Read Datasets and load your tables in database

Read the datasets into R dataframes using the urls provided above. Then load your tables in database.


```r
# Write your query here
# read in crop data
cropdf <- read.csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-RP0203EN-SkillsNetwork/labs/Final%20Project/Annual_Crop_Data.csv')
# load crop data into CROP_DATA table
dbWriteTable(conn, "CROP_DATA", cropdf, overwrite=TRUE, header=TRUE)

# read in farm price data
pricedf <- read.csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-RP0203EN-SkillsNetwork/labs/Final%20Project/Monthly_Farm_Prices.csv')
# load farm price data into FARM_PRICES table
dbWriteTable(conn, "FARM_PRICES", pricedf, overwrite=TRUE, header=TRUE)

# read in daily fx data
dailydf <- read.csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-RP0203EN-SkillsNetwork/labs/Final%20Project/Daily_FX.csv')
# load daily fx data into DAILY_FX table
dbWriteTable(conn, "DAILY_FX", dailydf, overwrite=TRUE, header=TRUE)

# read in monthly fx data
monthdf <- read.csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-RP0203EN-SkillsNetwork/labs/Final%20Project/Monthly_FX.csv')
# load monthly fx data into MONTHLY_FX table
dbWriteTable(conn, "MONTHLY_FX", monthdf, overwrite=TRUE, header=TRUE)

# print first few entries of each dataframe
head(cropdf)
head(pricedf)
head(dailydf)
head(monthdf)
```

## Now execute SQL queries using the RSQLite R package to solve the assignment problems.

## Problem 3

#### How many records are in the farm prices dataset?


```r
# Write your query here
dbGetQuery(conn, "SELECT COUNT(*) AS 'Record Count' FROM FARM_PRICES")
```

## Problem 4

#### Which geographies are included in the farm prices dataset?


```r
# Write your query here
dbGetQuery(conn, "SELECT DISTINCT GEO AS 'Record Locations' FROM FARM_PRICES")
```

## Problem 5

#### How many hectares of Rye were harvested in Canada in 1968?


```r
# Write your query here
dbGetQuery(conn, "SELECT SUM(HARVESTED_AREA) AS 'Hectares of Rye harvested' FROM CROP_DATA WHERE CROP_TYPE = 'Rye' AND GEO = 'Canada' AND SUBSTR(YEAR,1,4) = '1968'") # checking the year part of the YEAR date datatype
```

## Problem 6

#### Query and display the first 6 rows of the farm prices table for Rye.


```r
# Wriet your query here
dbGetQuery(conn, "SELECT * FROM FARM_PRICES WHERE CROP_TYPE = 'Rye' LIMIT 6")
```

## Problem 7

#### Which provinces grew Barley?


```r
# Write your query here
dbGetQuery(conn, "SELECT DISTINCT GEO AS 'Grew Barley' FROM CROP_DATA WHERE CROP_TYPE = 'Barley'")
```

## Problem 8

#### Find the first and last dates for the farm prices data.


```r
# Write your query here
dbGetQuery(conn, "SELECT MAX(DATE) AS 'Latest date', MIN(DATE) AS 'Earliest date' FROM FARM_PRICES")
```

## Problem 9

#### Which crops have ever reached a farm price greater than or equal to $350 per metric tonne?


```r
# Write your query here
dbGetQuery(conn, "SELECT DISTINCT CROP_TYPE AS '$350+ per Metric Ton' FROM FARM_PRICES WHERE PRICE_PRERMT >= 350")
```

## Problem 10

#### Rank the crop types harvested in Saskatchewan in the year 2000 by their average yield. Which crop performed best?


```r
# Write your query here
dbGetQuery(conn, "SELECT DISTINCT CROP_TYPE, MAX(AVG_YIELD) AS 'Maximum Average Yield in 2000' FROM CROP_DATA WHERE GEO = 'Saskatchewan' AND SUBSTR(YEAR,1,4) = '2000'
GROUP BY CROP_TYPE ORDER BY MAX(AVG_YIELD) DESC")
# Barley performed best in Saskatchewan in the year 2000
```

## Problem 11

#### Rank the crops and geographies by their average yield (KG per hectare) since the year 2000. Which crop and province had the highest average yield since the year 2000?


```r
# Write your query here
dbGetQuery(conn, "SELECT CROP_TYPE, GEO, AVG_YIELD FROM CROP_DATA WHERE SUBSTR(YEAR,1,4) > 2000
ORDER BY AVG_YIELD DESC LIMIT 5")
# Barley grown in Alberta had the highest average yield of 4100 since the year 2000
```

## Problem 12

#### Use a subquery to determine how much wheat was harvested in Canada in the most recent year of the data.


```r
# Write your query here
dbGetQuery(conn, "SELECT MAX(SUBSTR(YEAR, 1, 4)) AS 'Most recent year', SUM(HARVESTED_AREA) AS 'Total wheat harvested in Canada during year' FROM CROP_DATA
WHERE GEO = 'Canada' AND CROP_TYPE = 'Wheat' AND SUBSTR(YEAR, 1, 4) = (SELECT MAX(SUBSTR(YEAR, 1, 4)) FROM CROP_DATA)") # subquery to match result set to the latest year in the dataset
```

## Problem 13

#### Use an implicit inner join to calculate the monthly price per metric tonne of Canola grown in Saskatchewan in both Canadian and US dollars. Display the most recent 6 months of the data.


```r
# Write your query here
dbGetQuery(conn, "SELECT f.DATE, f.GEO, f.CROP_TYPE, f.PRICE_PRERMT AS 'Price per MT - CAD', (f.PRICE_PRERMT / m.FXUSDCAD) AS 'Price per MT - USD' FROM FARM_PRICES f , MONTHLY_FX m 
    WHERE SUBSTR(f.DATE, 1, 4) = SUBSTR(m.DATE, 1, 4) AND SUBSTR(f.DATE, 6, 2) = SUBSTR(m.DATE, 6, 2) AND CROP_TYPE = 'Canola' AND GEO = 'Saskatchewan'
    ORDER BY f.DATE DESC LIMIT 6") # substr functions to break DATE values into Year and Month (respectively)
```
